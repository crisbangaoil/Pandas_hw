```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
import pymongo
import os
import time
from splinter import Browser
import pandas as pd

# Convert Jupyter to Scrape function


# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define Database and collection
db = client.project_info
collection = db.project_info

executable_path = {"executable_path": "chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)

# #Scrape Monster
# url = 'https://www.monster.com/jobs/search/?q=data-science&where=San-Diego__2C-CA&intcid=skr_navigation_nhpso_searchMain'
# #browser
# browser.visit(url)
# #define the html
# time.sleep(1)
# #define the soup
# html=browser.html
# soup=bs(html,'html.parser')
# # Retrieve page with requests module
# response = requests.get(url)

# #Create BeautifulSOup object; parse with 'lxml'

# soup = bs(response.text,'lxml')

# # Retrieve data from Nasa website
# titles=soup.find_all('h2',class_='title')
title_list = []
# companies=soup.find_all('div',class_='company')
company_list = []
# locations= soup.find_all('a', class_='name')
location_list = []
# joburls = soup.find_all('h2', class_='title')
url_list = []

# for title in titles:
#     title_list.append(title.text)
#     url_list.append(title.a.get('href'))

# for location in locations:
#     location_list.append(location.text)

# for company in companies:
#     company_list.append(company.text)


# Scrape Career Builder
url = 'https://www.careerbuilder.com/jobs-data-science-in-san-diego,ca?keywords=data+science&location=San+Diego%2C+CA'
#browser
browser.visit(url)
#define the html
time.sleep(1)
#define the soup
html=browser.html
soup=bs(html,'html.parser')
# Retrieve page with requests module
response = requests.get(url)

#Create BeautifulSOup object; parse with 'lxml'

soup = bs(response.text,'lxml')

titles=soup.find_all('h2',class_='job-title hide-for-medium-up')
companies=soup.find_all('div', class_='columns large-2 medium-3 small-12')
locations= soup.find_all('div', class_='columns end large-2 medium-3 small-12')
joburls = soup.find_all('h2', class_='job-title hide-for-medium-up')

for title in titles:
    title_list.append(title.text)
    url_list.append('https://www.careerbuilder.com/'+title.a.get('href'))

for location in locations:
    location_list.append(location.text)

for company in companies:
    company_list.append(company.text)



#Scrape Indeed
url = 'https://www.indeed.com/jobs?q=data+science&l=San+Diego%2C+CA'

# Retrieve page with the requests module
response = requests.get(url)
soup = bs(response.text, 'lxml')
results = soup.find_all('div',class_='row')

# Loop through results to retrieve title, company, location, salary, and link
for result in results:

    #job title
    try:
        job_title = result.a.text.strip()
    except:
        job_title = 'empty'
    title_list.append(job_title)

    #company
    try:
        company = result.find(name='span', attrs={'class':'company'}).text.strip()
    except:
        company = 'empty'
    company_list.append(company)

    #location
    try:
        location = result.find(name='div', class_='location').text.strip() 
    except:
        try:
            location = result.find(name='span', class_='location').text.strip()
        except:
            location = 'no location'

    location_list.append(location)

    #salary
    # try:
    #     salary = result.find(name='span', class_='salary no-wrap').text.strip()
    # except:
    #     salary = 'no salary info'

    # salaries.append(salary)

    try:
        targets = result.find('a', class_='turnstileLink')
        link = 'https://www.indeed.com' + targets.attrs['href']
    except:
        link = 'empty'
    url_list.append(link)

#Scerpe Glassdoor
# request page content and create bs object
#response = requests.get(url)
#soup = bs(response.text,'lxml')
url = 'https://www.glassdoor.com/Job/jobs.htm?suggestCount=0&suggestChosen=false&clickSource=searchBtn&typedKeyword=data+science&sc.keyword=data+science&locT=C&locId=1147311&jobType='
browser.visit(url)
soup = bs(browser.html, "html.parser")

# On the result page
# Scrape
# Create list and append info

# Scrape title and url
for x in soup.find_all('div', class_="flexbox jobTitle"):
    try:
        title = x.text
        url = "https://www.glassdoor.com" + x.a.get("href")
    # skip and fill in "na" when tag has no text or link
    except:
        title = "NaN"

    title_list.append(title)
    url_list.append(url)

# Scrape company name    
for x in soup.find_all('div', class_="flexbox empLoc"):
    try:
        company = x.text

    # skip and fill in "na" when tag has no text or link
    except:
        company = "NaN"

    company_list.append(company)

# Scrape location
for x in soup.find_all('span', class_="subtle loc"):
    try:
        location = x.text

    # skip and fill in "na" when tag has no text or link
    except:
        location = "NaN"

    location_list.append(location)

#Create dictionary to hold all information
    posts = {
        'title': title_list,
        'location':location_list,
        'company': company_list,
        'url':url_list
        }   

#convert dictionary to Pandas Data Frame
df = pd.DataFrame.from_dict(posts)

#drop duplicates of title and location are the same
df = df.drop_duplicates(subset=['title','location','company'],keep='last')

#convert data_frame back to dictionary
records = df.to_dict('records')

#insert into each listing into DB
for record in records:
    collection.insert_one(record)
    
records
```
