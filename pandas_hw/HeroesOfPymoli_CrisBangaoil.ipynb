### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data["Purchase ID"].count()
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unumber_items = purchase_data["Item Name"].nunique()
average_price = purchase_data["Price"].mean()
total_revenue = purchase_data["Price"].sum()



summary = pd.DataFrame({"Number of Unique Items": [unumber_items],"Average Price": [average_price], "Number of Purchases": [total_players], "Total Revenue": [total_revenue]})
summary["Average Price"] = summary["Average Price"].map("${:.2f}".format)
summary["Total Revenue"] = summary["Total Revenue"].map("${:.2f}".format)
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group = purchase_data.groupby(["Gender"])
gender_count = gender_group.count()
gender_count["Percentage"] = gender_count["Purchase ID"] / total_players * 100
gender_count["Percentage"] = gender_count["Percentage"].map("{:.2f}%".format)
gender_summary = gender_count.loc[:,["Percentage"]]
gender_summary

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_group = purchase_data.groupby(["Gender"])
gender_avg = gender_group.mean()
gender_avg["Purchase Count"] = gender_group["Purchase ID"].count()
gender_avg["Average Purchase Price"] = gender_group["Price"].mean()
gender_avg["Total Purchase Value"] = gender_group["Price"].sum()
gender_avg["Avg Total Purchase per Person"] = gender_avg["Total Purchase Value"] / gender_group["SN"].nunique() 

gender_avg["Average Purchase Price"] = gender_avg["Average Purchase Price"].map("${:.2f}".format)
gender_avg["Total Purchase Value"] = gender_avg["Total Purchase Value"].map("${:.2f}".format)
gender_avg["Avg Total Purchase per Person"] = gender_avg["Avg Total Purchase per Person"].map("${:.2f}".format)

gender_avg = gender_avg.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]



gender_avg


```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
max_age = purchase_data["Age"].max()
bins = [0, 9, 14, 19, 24, 29, 34, 39, max_age]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]

purchase_data["Age Summary"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)
age_group = purchase_data.groupby(["Age Summary"])
age_summary = age_group.count() 
age_percentage = age_summary.rename(columns={"Purchase ID":"Total Count"})
age_percentage["Percentage of Players"] = age_percentage["Total Count"] / total_players * 100
age_percentage["Percentage of Players"] = age_percentage["Percentage of Players"].map("{:.2f}%".format)
age_percentage = age_percentage.loc[:,["Total Count", "Percentage of Players"]]
age_percentage
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_count = age_group["Purchase ID"].count()
age_avg_price = age_group["Price"].mean()
age_total_purchase = age_group["Price"].sum()
age_unique_players = age_group["Price"].sum() / age_group["SN"].nunique()
age_summary_table = pd.DataFrame({"Purchase Count": age_count, "Average Purchase Price": age_avg_price, "Total Purchase Value": age_total_purchase, "Avg Total Purchase per Person": age_unique_players})
age_summary_table["Average Purchase Price"] = age_summary_table["Average Purchase Price"].map("${:.2f}".format)
age_summary_table["Total Purchase Value"] = age_summary_table["Total Purchase Value"].map("${:.2f}".format)
age_summary_table["Avg Total Purchase per Person"] = age_summary_table["Avg Total Purchase per Person"].map("${:.2f}".format)
age_summary_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender = purchase_data.groupby(["SN"])
spender_count = spender["Purchase ID"].count()
spender_avg = spender["Price"].mean()
spender_total = spender["Price"].sum()
spender_summary = pd.DataFrame({"Purchase Count": spender_count, "Average Purchase Price": spender_avg, "Total Purchase Value": spender_total})
spender_sort = spender_summary.sort_values("Total Purchase Value", ascending=False) 
spender_sort["Average Purchase Price"] = spender_summary["Average Purchase Price"].map("${:.2f}".format)
spender_sort["Total Purchase Value"] = spender_summary["Total Purchase Value"].map("${:.2f}".format)                                                       
spender_sort.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item = purchase_data.groupby(["Item ID"])
purchase_count = item["Purchase ID"].count()
item_price = item["Price"].max()
total = item["Price"].sum()
item_name = item["Item Name"].max()
high_item = pd.DataFrame({"Item Name":item_name, "Purchase Count": purchase_count, "Item Price": item_price, "Total Purchase Value": total})
high_item_sort = high_item.sort_values("Purchase Count", ascending=False)
high_item_sort["Item Price"] = high_item_sort["Item Price"].map("${:.2f}".format)
high_item_sort["Total Purchase Value"] = high_item_sort["Total Purchase Value"].map("${:.2f}".format)

high_item_sort.head()
```
